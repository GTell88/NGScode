# Merge dataframes

```python
merged_df = pd.merge(GDS_df, seq_df, 
                  left_on = ['Patient','Visit'],
                  right_on = ['Patient','Visit'],
                  how = 'inner')

merged_df.sort_values(['AAPos','Patient','Visit'],  inplace=True)
merged_df.head()
```
